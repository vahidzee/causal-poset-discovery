# Estimating Interventional Distributions

You can use this notebook to estimate the interventional distributions using a trained model. In order to reproduce the training of the models, you can do as follows:

1. Set up your virtual environment and install the requirements.
2. From the root of the repository, run the following command:
    - To reproduce the main results in the paper:
        ```bash
        python trainer.py fit \
            --config experiments/intervention/config/main.yml \
            --data experiments/intervention/data/main.yml
        ```
    - To reproduce the results in the appendix:
        ```bash
        python trainer.py fit \ 
            --config experiments/intervention/config/supplements.yml \
            --data experiments/intervention/data/supplements.yml
        ```
3. Choose the checkpoint that you want to use for the estimation as the appropriate checkpoint path in the rest of this notebook.
